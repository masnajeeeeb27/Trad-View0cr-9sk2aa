```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("processed.va.data",header=None)
```

```python
header = ["age","sex","cp","trestbps","chol","fbs","restecg",
          "thalach","exang","oldpeak","slope","ca","thal","class_att"]
```

```python
df.columns = header
df.head(10)
```

```python
df = df.replace('?',np.NaN)
df
```

```python
# melakukan oversampling pada nilai yang nan
df.isnull().sum()
```

```python
# menghapus kolom slope, ca,thal
kolom_yang_dihapus = ['slope', 'ca', 'thal']
df = df.drop(kolom_yang_dihapus, axis=1)
df.isnull().sum()
```

```python
df.dtypes
```

```python
# Mengonversi kolom 'trestbps' ke tipe data numerik
df['trestbps'] = pd.to_numeric(df['trestbps'], errors='coerce')

# Menghitung rata-rata kolom 'trestbps' tanpa memasukkan nilai NaN
avg_trestbps = df['trestbps'].mean(skipna=True)

# Mengisi NaN values dengan rata-rata
df['trestbps'].fillna(avg_trestbps, inplace=True)

# Mengonversi kolom 'trestbps' ke tipe data integer
df['trestbps'] = df['trestbps'].astype('int64')

# Menampilkan dataframe setelah konversi dan pengisian nilai NaN
print("\nDataframe setelah konversi dan pengisian nilai NaN:")
print(df)
```

```python
# Mengonversi kolom 'trestbps' ke tipe data numerik
df['chol'] = pd.to_numeric(df['chol'], errors='coerce')

# Menghitung rata-rata kolom 'trestbps' tanpa memasukkan nilai NaN
avg_trestbps = df['chol'].mean(skipna=True)

# Mengisi NaN values dengan rata-rata
df['chol'].fillna(avg_trestbps, inplace=True)

# Mengonversi kolom 'trestbps' ke tipe data integer
df['chol'] = df['chol'].astype('int64')
```

```python
df['fbs'].value_counts().idxmax()
df['fbs'].replace(np.nan, 0, inplace=True)
```

```python
# Mengonversi kolom 'trestbps' ke tipe data numerik
df['thalach'] = pd.to_numeric(df['thalach'], errors='coerce')

# Menghitung rata-rata kolom 'trestbps' tanpa memasukkan nilai NaN
avg_trestbps = df['thalach'].mean(skipna=True)

# Mengisi NaN values dengan rata-rata
df['thalach'].fillna(avg_trestbps, inplace=True)

# Mengonversi kolom 'trestbps' ke tipe data integer
df['thalach'] = df['thalach'].astype('int64')

```

```python
df['exang'].value_counts().idxmax()
df['exang'].replace(np.nan, 140, inplace=True)
```

```python
df['oldpeak'].value_counts().idxmax()
df['oldpeak'].replace(np.nan, 140, inplace=True)
```

```python
fitur = df.iloc[:, :-1]
target = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
# Melakukan pembagian data menjadi data latih dan data uji
# Proporsi data latih dapat diatur dengan parameter test_size
# Dalam contoh ini, 70% data akan digunakan sebagai data latih (train)
X_train, X_test, y_train, y_test = train_test_split(fitur, target, test_size=0.3, random_state=42)

# Menampilkan ukuran data latih dan data uji
print("Ukuran data latih (train):", X_train.shape)
print("Ukuran data uji (test):", X_test.shape)
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score


# Oversampling dengan RandomOverSampler
oversampler = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_resampled_over, y_resampled_over = oversampler.fit_resample(X_train, y_train)

# Undersampling dengan RandomUnderSampler
undersampler = RandomUnderSampler(sampling_strategy='auto', random_state=42)
X_resampled_under, y_resampled_under = undersampler.fit_resample(X_train, y_train)

# Membuat model SVM
svm_model = SVC()

# Hyperparameter tuning menggunakan GridSearchCV pada oversampled data
param_grid = {'C': [0.1, 1, 10], 'kernel': ['linear', 'rbf']}
grid_search_over = GridSearchCV(svm_model, param_grid, scoring='accuracy', cv=5)
grid_search_over.fit(X_resampled_over, y_resampled_over)

# Hyperparameter tuning menggunakan GridSearchCV pada undersampled data
grid_search_under = GridSearchCV(svm_model, param_grid, scoring='accuracy', cv=5)
grid_search_under.fit(X_resampled_under, y_resampled_under)

# Menggunakan model terbaik setelah hyperparameter tuning pada oversampled data
best_svm_over = grid_search_over.best_estimator_

# Menggunakan model terbaik setelah hyperparameter tuning pada undersampled data
best_svm_under = grid_search_under.best_estimator_

# Memprediksi kelas target pada data uji
y_pred_over = best_svm_over.predict(X_test)
y_pred_under = best_svm_under.predict(X_test)

# Mengukur akurasi model pada data uji
accuracy_over = accuracy_score(y_test, y_pred_over)
accuracy_under = accuracy_score(y_test, y_pred_under)

print("Akurasi Model SVM dengan Oversampling:", accuracy_over)
print("Akurasi Model SVM dengan Undersampling:", accuracy_under)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd



# Oversampling dengan RandomOverSampler
oversampler = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_resampled_over, y_resampled_over = oversampler.fit_resample(X_train, y_train)

# Undersampling dengan RandomUnderSampler
undersampler = RandomUnderSampler(sampling_strategy='auto', random_state=42)
X_resampled_under, y_resampled_under = undersampler.fit_resample(X_train, y_train)

# Hyperparameter tuning untuk KNN
param_grid_knn = {'n_neighbors': [3, 5, 7], 'weights': ['uniform', 'distance']}
knn_model = KNeighborsClassifier()
grid_search_knn_over = GridSearchCV(knn_model, param_grid_knn, scoring='accuracy', cv=5)
grid_search_knn_under = GridSearchCV(knn_model, param_grid_knn, scoring='accuracy', cv=5)

# Hyperparameter tuning untuk Decision Tree
param_grid_dt = {'max_depth': [3, 5, 7], 'criterion': ['gini', 'entropy']}
dt_model = DecisionTreeClassifier()
grid_search_dt_over = GridSearchCV(dt_model, param_grid_dt, scoring='accuracy', cv=5)
grid_search_dt_under = GridSearchCV(dt_model, param_grid_dt, scoring='accuracy', cv=5)

# Melatih model KNN dengan data oversampled dan undersampled
grid_search_knn_over.fit(X_resampled_over, y_resampled_over)
grid_search_knn_under.fit(X_resampled_under, y_resampled_under)

# Melatih model Decision Tree dengan data oversampled dan undersampled
grid_search_dt_over.fit(X_resampled_over, y_resampled_over)
grid_search_dt_under.fit(X_resampled_under, y_resampled_under)

# Menggunakan model terbaik setelah hyperparameter tuning pada KNN
best_knn_over = grid_search_knn_over.best_estimator_
best_knn_under = grid_search_knn_under.best_estimator_

# Menggunakan model terbaik setelah hyperparameter tuning pada Decision Tree
best_dt_over = grid_search_dt_over.best_estimator_
best_dt_under = grid_search_dt_under.best_estimator_

# Memprediksi kelas target pada data uji menggunakan KNN
y_pred_knn_over = best_knn_over.predict(X_test)
y_pred_knn_under = best_knn_under.predict(X_test)

# Memprediksi kelas target pada data uji menggunakan Decision Tree
y_pred_dt_over = best_dt_over.predict(X_test)
y_pred_dt_under = best_dt_under.predict(X_test)

# Mengukur akurasi model pada data uji untuk KNN
accuracy_knn_over = accuracy_score(y_test, y_pred_knn_over)
accuracy_knn_under = accuracy_score(y_test, y_pred_knn_under)

# Mengukur akurasi model pada data uji untuk Decision Tree
accuracy_dt_over = accuracy_score(y_test, y_pred_dt_over)
accuracy_dt_under = accuracy_score(y_test, y_pred_dt_under)

print("Akurasi Model KNN dengan Oversampling:", accuracy_knn_over)
print("Akurasi Model KNN dengan Undersampling:", accuracy_knn_under)

print("Akurasi Model Decision Tree dengan Oversampling:", accuracy_dt_over)
print("Akurasi Model Decision Tree dengan Undersampling:", accuracy_dt_under)
```
